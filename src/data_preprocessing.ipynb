```python
df = pd.DataFrame({'date':date})
df
```

```python
import datetime
df['dt'] = pd.to_datetime(df['date'])
df['quarter'] = df['dt'].dt.year.apply(str) + 'Q' + df['dt'].dt.quarter.apply(str)
df['dt'] = df['dt'].dt.strftime('%Y-%m')
df
```

```python
KOR_price['dt'] = pd.to_datetime(KOR_price['index'])
KOR_price['dt'] = KOR_price['dt'].dt.strftime('%Y-%m')
JP_price['dt'] = pd.to_datetime(JP_price['index'])
JP_price['dt'] = JP_price['dt'].dt.strftime('%Y-%m')
JP_price[4][278] = 101.1
KOR_price, JP_price
```

```python
KOR_price_list = KOR_price.iloc[:,1].to_list()
KOR_index_list = KOR_price['dt'].to_list()
JP_price_list = JP_price.iloc[:,1].to_list()
JP_index_list = JP_price['dt'].to_list()
df['KOR_price'] = 'nan'
df['JP_price'] = 'nan'
for i in range(len(df)):
    try:
        df['KOR_price'][i] = KOR_price_list[KOR_index_list.index(df.iloc[i,1])]
        df['JP_price'][i] = JP_price_list[JP_index_list.index(df.iloc[i,1])]
    except: pass
df.head(50)
```
